```python
import statsmodels.api as sm
from pylab import rcParams
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(Daily2PACO_T1L1, model='additive')
fig = decomposition.plot()
plt.show()
```

```python
import itertools
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
warnings.filterwarnings("ignore")
```

```python
Daily2PACO_T1L1
```

```python
aic_val = []
param_val = []
param_seasonal_val = []

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(Daily2PACO_T1L1,order=param,seasonal_order=param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
            aic_val.append(results.aic)
            param_val.append(param)
            param_seasonal_val.append(param_seasonal)
        except:
            continue
```

```python
#param_seasonal_val
```

```python
param_and_seasonal = [i for tup in zip(param_val,param_seasonal_val) for i in tup]
```

```python
combine_param_and_seasonal = [a+b for a, b in zip(param_and_seasonal[::2],param_and_seasonal[1::2])]
```

```python
#combine_param_and_seasonal
```

```python
aic_val_and_combined = [i for tup in zip(aic_val,combine_param_and_seasonal) for i in tup]
```

```python
#aic_val_and_combined
```

```python
def to_dictionary(lst):
    res_dct = {lst[i]: lst[i + 1] for i in range(0, len(lst), 2)}
    return res_dct
```

```python
pdq_aic_val = to_dictionary(aic_val_and_combined)
```

```python
optimal_order = pdq_aic_val[min(aic_val)]
```
